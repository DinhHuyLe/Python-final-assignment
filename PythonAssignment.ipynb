```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://ir.tesla.com/press-release/tesla-third-quarter-2024-production-deliveries-and-deployments'
response = requests.get(url)

BeauSoup = BeautifulSoup(response.text, 'html.parser')
tables = BeauSoup.find_all('table')
tesla_revenue = pd.read_html(str(tables))[0]

tesla_revenue.tail()
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()

make_graph(tesla_share_price_data, 'Tesla Stock Price')
```
